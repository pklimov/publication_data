# Raw $T_1$ data for "Fluctuations of Energy-Relaxation Times in Superconducting Qubits"

Letter:
- [Phys Rev Letters 121, 090502 (2018)](https://storage.googleapis.com/gweb-research2023-media/pubtools/pdf/5dc59b7f00e2338d1888ab52df771a7cf72fd5b6.pdf)
- [arXiv:1809.01043 (2018)](https://arxiv.org/pdf/1809.01043.pdf)

```python
import os

import matplotlib.pyplot as plt
import pandas as pd
```

```python
DIRECTORY = os.getcwd()
DIRECTORY
```

```python
QUBITS = ("q1_3", "q1_2", "q0_1", "q0_3", "q0_4")

print('Will attempt to pull files from:')
for qubit in QUBITS:
    QUBIT_DIRECTORY = f'{DIRECTORY}/{qubit}/'
    print(f'Pulling files from {QUBIT_DIRECTORY}')
```

```python
fig, axes = plt.subplots(len(QUBITS), 1, sharex=True, figsize=(10, 10))

for axis, qubit in zip(axes, QUBITS):
    QUBIT_DIRECTORY = f'{DIRECTORY}/{qubit}/'

    freqs_GHz_mesh = pd.read_csv(QUBIT_DIRECTORY + 'freqs_GHz_mesh.csv')
    times_min_mesh = pd.read_csv(QUBIT_DIRECTORY + 'times_min_mesh.csv')
    T1s_us_mesh = pd.read_csv(QUBIT_DIRECTORY + 'T1s_us_mesh.csv')

    cplt = axis.pcolormesh(
        freqs_GHz_mesh,
        times_min_mesh,
        T1s_us_mesh,
        cmap='RdYlBu',
        vmin=0,
        vmax=48,
    )
    axis.set_ylabel('Time (minutes)')
    axis.set_xlabel('Frequency (GHz)')
    axis.set_title(f'qubit: {qubit}')
    plt.colorbar(cplt, ax=axis, label=r'$T_1$ ($\mu$s)')

plt.tight_layout()
```
